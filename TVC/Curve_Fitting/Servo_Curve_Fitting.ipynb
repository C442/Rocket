```python
input_file = "sim_output_new.csv"
output_file = "filtered_data3.csv"

def is_multiple_of_0_01(t):
    return abs(round(t * 100) - t * 100) < 1e-6  # avoids float errors

with open(input_file, 'r') as fin, open(output_file, 'w') as fout:
    for line in fin:
        line = line.strip()
        if not line:
            continue  # skip empty lines

        parts = line.split(',')
        try:
            t = float(parts[0])
            if is_multiple_of_0_01(t):
                fout.write(line + '\n')
        except ValueError:
            # Possibly header or invalid row — keep it if needed
            if not parts[0].replace('.', '', 1).isdigit():
                fout.write(line + '\n')  # optionally keep header
```

```python
import math

# Constants (in mm)
l1_s1 = 10.5    # length of servo attach
l2_s1 = 32.5   # bar length
l3_s1 = 42.5     # distance from pin to bar on the 3D printed part
Cx_s1 = 30.0    # x distance between servo wheel and mount pin
Cy_s1 = -29.0  # y distance

import numpy as np

def m_to_s1(a):
    a = np.radians(90-a)
    
    px = Cx_s1 + l3_s1 * np.cos(a)
    py = Cy_s1 + l3_s1 * np.sin(a)
    d = np.sqrt(px**2 + py**2)

    x1 = (l1_s1**2 + d**2 - l2_s1**2) / (2 * d)
    y1 = np.sqrt(l1_s1**2 - x1**2)

    ang = np.arctan2(py, px)

    x2 = x1 * np.cos(ang) - y1 * np.sin(ang)
    y2 = x1 * np.sin(ang) + y1 * np.cos(ang)


    return np.degrees(np.arctan2(y2, x2))

```

```python
gimbal_angles = np.arange(-5,5.1,0.1)
res = m_to_s1(gimbal_angles)
fig, ax = plt.subplots(figsize = (15,6))
ax.plot(gimbal_angles, res)
ax.set_xlabel("Gimbal Angle (deg)")
ax.set_ylabel("Servo Angle (deg)")
```
